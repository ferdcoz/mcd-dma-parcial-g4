## 3. Las N-Reinas

En un hipotético tablero de ajedrez n x n, posicionar n reinas, de tal manera que ninguna ataque al resto. 

Se resuelve usando la técnica de backtracking. El ejemplo de uso en el código resuelve el problema de las N reinas (n = 8) y muestra las primeras 3 soluciones.

```python
import matplotlib.pyplot as plt

def es_posicion_segura(tablero, fila, columna):
    for i in range(fila):
        if tablero[i] == columna or \
           tablero[i] - i == columna - fila or \
           tablero[i] + i == columna + fila:
            return False
    return True

def resolver_n_reinas(n):
    tablero = [-1] * n
    soluciones = []

    def resolver_util(fila):
        if fila == n:
            # Si se encuentra una solución, agregar
            soluciones.append(tablero[:])
            return

        for columna in range(n):
            if es_posicion_segura(tablero, fila, columna):
                tablero[fila] = columna
                resolver_util(fila + 1)
                tablero[fila] = -1

    resolver_util(0)

    return soluciones

def dibujar_tablero(tablero):
    n = len(tablero)

    plt.figure(figsize=(n, n))
    plt.axis('off')
    for fila in range(n):
        for columna in range(n):
            color = 'white' if (fila + columna) % 2 == 0 else 'lightgray'
            plt.fill([columna, columna, columna + 1, columna + 1],
                     [n - fila - 1, n - fila, n - fila, n - fila - 1],
                     color)

    for fila, columna in enumerate(tablero):
        plt.text(columna + 0.5, n - fila - 0.5, '♛', fontsize=30,
                 color='black', ha='center', va='center')

    plt.show()

n = 8
soluciones = resolver_n_reinas(n)

for i, solucion in enumerate(soluciones[:3]):
    print(f"Solución {i+1}:")
    dibujar_tablero(solucion)
```
